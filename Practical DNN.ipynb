```python
%%plantuml
@startuml

floating note right
No bias varaince tradeoff any more, if we can make NN bigger 
and at same time train more data with proper regularization, 
we can decrease bias and varaince.
end note

repeat
while(high bias?) is (Yes)
    :bigger network;
    :train longer;
endwhile (No)        
while(high variance?) is (Yes)
    :More Data;
    :Regularization;
endwhile (No)
repeat while (high bias?) is (yes)
stop

@enduml
```

### Data set is split into three sets:
1. Training Set
2. Cross Validation Set, hold out set, development Set
3. Test Set

### Split of Dataset
- Small/Medium Data Set: 60/20/20
- Big Data Set: 98/1/1, even 99.8/0.1/0.1

### Bias, Variance, underfitting, overfitting
- Underfitting: high training error => High Bias
- Overfitting: high test/validation error => High Variance

### Reduce Bias/Underfitting
- Bigger Network
- Traning Longer

### Reduce Variance/Overfitting
- More Data
- Regularization
